##VV3. Fall Rosenbrocks

√ç √¶fingu √≠ kafla A3 er ‚ÄûFall Rosenbrocks‚Äú skilgreint og s√Ωnd mynd af √æv√≠. Falli√∞ er:

$ùëì(ùë•,ùë¶)=(1‚àíùë•)^2+100(ùë¶‚àíùë•^2)^2$



a. Skrifi√∞ fall `rosen(x,y)` til a√∞ reikna gildi fallsins. Pr√≥fi√∞ a√∞ reikna ùëì(‚àí1.2,1) sem √¶tti a√∞ gefa 24.2.</li>



```python
#a

def rosen(x,y):
  a = ((1-x)**2 + 100*(y-x**2)**2)
  return a

print(f'{rosen(-1.2,1):.1f}')
```

b. Teikni√∞ h√¶√∞arl√≠numynd af Rosenbrock-fallinu me√∞ `plt.contour` √° sv√¶√∞inu [‚àí2,2]√ó[‚àí1,3]. L√°ti√∞ h√¶√∞arl√≠nurnar vera d√∂kkbl√°ar, merki√∞ √æ√¶r l√≠kt og gert er √≠ s√Ωnid√¶minu H√¶√∞arl√≠nur falls √≠ kafla A1, og hafi√∞ netl√≠nur (grid) me√∞. Til a√∞ f√° g√≥√∞a mynd √æarf b√¶√∞i a√∞ reikna falli√∞ √° √æ√©ttu neti (t.d. 300√ó300) og velja h√¶√∞arl√≠nugildin, sem f√°st me√∞ stikanum `levels` vel. Gildin 0.2,22,62,102,‚Ä¶,462 gefa t.d. √°g√¶ta mynd. √ûau m√° b√∫a til me√∞: `levels = np.append([0.2], np.arange(2,50,4)**2)`

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-2,2,300)
y = np.linspace(-1,3,300)
(X,Y) = np.meshgrid(x,y)
Z = rosen(X,Y)

levels = np.append([0.2], np.arange(2,50,4)**2)
c = plt.contour(X,Y,Z,levels,colors='darkblue')
plt.clabel(c,fmt='%.0f')
plt.grid()
plt.show()
```
