Hér er reitur með (hluta af) lausn á æfingunni í kafla 13.1, sem metur með hermun líkurnar á því að fá tvær tvennur í póker. Fallið pókerhönd er endurbætt miðað við æfinguna til að tryggja að sama spil geti ekki komið fyrir tvisvar. Fræðilegar líkur á tveimur tvennum er 4.75% og ef keyrt er með háu gildi á n fæst hlutfall nálægt því.

```
from random import sample
```

Í lausninni er spil er táknað með pari `(sort, gildi)` þar sem sortin er á bilinu 1–4 (hjarta, spaði, tígull, lauf) og gildi er á bilinu 1–13 (ás, tvistur,…, kóngur).

Skrifið föll sem kanna hvort hönd geymi:
<ol type="a">
  <li>tvennu</li>
  <li>þrennu</li>
  <li>fullt hús (3 + 2)</li>
  <li>lit (t.d. eintómir spaðar)</li>

</ol>

Hermið slíkar hendur og berið saman við raunverulegar líkur fengnar af netinu.


````python
#a
from random import sample

def pókerhönd():
  stokkur = [(i,j) for i in range(1,5) for j in range(1,14)]
  return sample(stokkur, 5)

def tvenna(hönd):
  gildi = [spil[1] for spil in hönd]
  M = set(gildi)                 # búa til mengi
  if len(M) != 4:
    return False                 # það verða að vera 4 gildi
  for spil in M:
    talning = gildi.count(spil)
    if talning > 2: return False # og ekkert spil má koma 3svar
  return True                    # það eru tvenna

def hlutfall_með_tvennu(n):
  talning = 0
  for i in range(n):
    hönd = pókerhönd()
    if tvenna(hönd):
      talning += 1
  return talning/n

hf = hlutfall_með_tvennu(100000)
print(f'Hlutfall með tvennu: {hf*100:.1f}%')
````

````python
#b

def þrenna(hönd):
  gildi = [spil[1] for spil in hönd]
  M = set(gildi)                 # búa til mengi
  if len(M) != 3:
    return False                 # það verða að vera 3 gildi
  for spil in M:
    talning = gildi.count(spil)
    if talning > 3: return False # og ekkert spil má koma 4sinnum
  return True                    # það er þrenna

def hlutfall_með_þrennu(n):
  talning = 0
  for i in range(n):
    hönd = pókerhönd()
    if þrenna(hönd):
      talning += 1
  return talning/n

hf = hlutfall_með_þrennu(100000)
print(f'Hlutfall með þrennu: {hf*100:.1f}%')
````

````python
#c

def fullthús(hönd):
  gildi = [spil[1] for spil in hönd]
  M = set(gildi)                 # búa til mengi
  if len(M) != 2:
    return False                 # það verða að vera 2 gildi
  for spil in M:
    talning = gildi.count(spil)
    if talning > 3: return False # og ekkert spil má koma 4 sinnum
  return True                    # það er fullt hús

def hlutfall_með_fullthús(n):
  talning = 0
  for i in range(n):
    hönd = pókerhönd()
    if fullthús(hönd):
      talning += 1
  return talning/n

hf = hlutfall_með_fullthús(100000)
print(f'Hlutfall með fullthús: {hf*100:.3f}%')
````

````python
#d

def litur(hönd):
  sort = [spil[0] for spil in hönd]
  gildi = [spil[1] for spil in hönd]
  N = set(sort)
  if len(N) != 1:                # það verður að vera 1 sort
    return False
  M = set(gildi)                 
  if len(M) != 5:
    return False                 # það verða að vera 5 gildi
  for spil in M:
    talning = gildi.count(spil)
    if talning > 1: return False # og ekkert spil má koma 2 sinnum
  return True                    

def hlutfall_með_litur(n):
  talning = 0
  for i in range(n):
    hönd = pókerhönd()
    if litur(hönd):
      talning += 1
  return talning/n

hf = hlutfall_með_litur(100000)
print(f'Hlutfall með litur: {hf*100:.3f}%')
````
